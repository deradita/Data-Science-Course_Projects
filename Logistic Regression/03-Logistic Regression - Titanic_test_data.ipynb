Firstly, let's import the necessary libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Next, we'll read in the Titanic_train dataset

```python
train_data = pd.read_csv('Titanic_train.csv')
train_data.head()
```

Moving further, we put the data through a bit of preprocessing

```python
train_data.drop('Cabin', axis = 1, inplace = True )
```

```python
train_data.dropna(inplace = True )
```

```python
sex = pd.get_dummies(train_data['Sex'], drop_first = True)
embark = pd.get_dummies(train_data['Embarked'], drop_first = True)
```

```python
train_data.drop(['Sex', 'Embarked', 'Name', 'Ticket'], axis = 1, inplace = True)
```

```python
train_data = pd.concat([train_data, sex, embark], axis = 1)
```

```python
train_data.head()
```

```python
test_data = pd.read_csv('Titanic_test.csv')
test_data 
```

```python
test_data.drop('Cabin', axis = 1, inplace = True )
```

```python
test_data.dropna(inplace = True )
```

```python
sex1 = pd.get_dummies(test_data['Sex'], drop_first = True)
embark1 = pd.get_dummies(test_data['Embarked'], drop_first = True)
```

```python
test_data.drop(['Sex', 'Embarked', 'Name', 'Ticket'], axis = 1, inplace = True)
```

```python
test_data = pd.concat([test_data, sex1, embark1], axis = 1)
```

```python
test_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y_test = test_data 
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_data.drop('Survived', axis = 1), train_data['Survived'], test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver = 'liblinear')
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
print(classification_report(y_test, predictions))
```
