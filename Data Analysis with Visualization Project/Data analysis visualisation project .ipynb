## Visualisation

Firstly, we need to obtain information about relationships in the dataset
In simpler terms, getting to know the variables that are in <b>correlation</b> with each other

```python
sns.pairplot(train)
```

To get a clearer and more numeric view of the relationships, we use a heatmap

```python
plt.subplots(figsize=(15,10))
sns.heatmap(train.corr(), annot = True, fmt = '1f')

#To get a clearer view of the heatmap
 
```

A complete correlation between two variables is expressed as <b>1</b>. <br>
From the visualisation above, we can tell that the highest correlation value is <b>0.50458</b> which approximately equals <b>1</b>. <br>
This value is expressed between <b>Age</b> and <b>Hypertension.</b><br>
So, we can say that there exists a relationship between <b>Age</b> and <b>Hypertension</b>;<br>
The higher the increase in <b>Age</b>, the higher the increase in <b>Hypertension</b> and vice versa.

With this knowledge, we can further analyze with respect to <b>Age</b> and <b>Hypertension</b> to gain further insight
    

So we further analyze that by using a countplot to discover the highest age group to suffer hypertension

Using a countplot, let us have a view of the highest age group to suffer from hypertension 

```python
sns.countplot(y = 'Hipertension', data = train, palette = 'cool')
```

The countplot above gives a result of over 8,000 individuals suffering from hipertension and less than 3,000 individuals free from hipertension<br> 
This tells us that the number of individuals suffering from hypertension is much more than the number of individuals that are not affected by it <br>


Now we determine the age grade that suffers the most from hypertension using a countplot

```python
sns.barplot(x = 'Hipertension', y = 'Age', data = train, palette = 'cool')
```

```python
sns.boxplot(x='Hipertension', y='Age', data=train, palette='rainbow')
```

From the countplot above, we now know that mostly individuals within the age grade of 30 - 95 and above suffer more from hipertension

```python
sns.boxplot(x='Hipertension', y='Age', hue='Gender', data=train, palette='coolwarm')
```

Furthermore, an approximately equal number of women and men in this age group suffer from hipertension. Therefore, the occurence of hipertension is non-gender-specific
